```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python

new_df = pd.read_csv("Data_after_DT.csv")
display(new_df.head())
display(new_df.info())
```

**VISUALIZATION**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Simple Actual and discounted Prices distribution visualization

```python
plt.figure(figsize= (10,5))
sns.histplot(new_df['actual_price'], bins=30,  color='blue', label="Actual Price")
sns.histplot(new_df['discounted_price'], bins=30, color='red', label=" Discounted Price")
plt.legend()
plt.title(" Actual & Discounted Prices distribution")
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.boxplot(x=new_df["actual_price"])
plt.show()
```

MinMax Scaling has scaled price between 0 to 1 there may be more outliers, that we can remove some extreme outliers..

```python

Q1 = new_df['actual_price'].quantile(0.25)
Q3 = new_df['actual_price'].quantile(0.75)


IQR = Q3 - Q1
print(IQR)

display(Q1)
display(Q3)

upper_bound = Q3 + 1.5 * IQR
new_df = new_df[new_df['actual_price'] <= upper_bound]
display("dataset after removing outlier:", new_df)




```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.histplot(new_df['actual_price'], bins=30, color='blue', label="Actual Price")
sns.histplot(new_df['discounted_price'], bins=30,  color='red', label="Discounted Price")
plt.legend()
plt.title("Distribution of Actual & Discounted Prices (After Outlier Removal)")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x=new_df['actual_price'])
plt.title("Boxplot of Actual Prices After Outlier Removal")
plt.show()
```

```python
#Average Discount Per Category
plt.figure(figsize=(12,6))

top_categories = new_df['category'].value_counts().nlargest(10).index
filtered_df = new_df[new_df['category'].isin(top_categories)]

sns.barplot(x=filtered_df['category'], y=filtered_df['discount_amount'])
plt.xticks(rotation=46, ha="right")
plt.title("Average Discount Amount per Category")
plt.xlabel("Category")
plt.ylabel("Average Discount")
plt.show()

plt.figure(figsize=(8, 9))
new_df['has_discount'].value_counts().plot.pie(autopct='%1.1f%%', colors=["blue", "red"])
plt.title("Percentage of Products with Discounts")
plt.ylabel("")
plt.show()
```

**Take away form the analysis**


*   Most products have discounts (96%) and (3%) have no discount
*   Electronics & accessories get the highest discounts, likely due to competitive pricing.
*   Outliers in price data were removed, making analysis more accurate.
